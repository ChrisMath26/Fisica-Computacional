# Integración y Derivación Numérica



Fuerza gravitacional producida por una hoja cuadrada uniforme: Un plato de metal cuadrado flota en el espacio como muestra la figura:

image.png

La hoja tiene $L=10$m de cada lado, un espesor despreciable y pesa 10 toneladas. Para este ejercicio es posible demostrar que la fuerza gravitacional que ejerce la hoja metálica sobre la masa puntal $m_1$ de 1kg a lo largo del eje $z$ es:

$$
F_z = G\sigma m_1 \int^{L/2}_{-L/2} \int^{L/2}_{-L/2} \frac{dx dy}{(x^2 + y^2 + z^2)}
$$

Aquí $G = 6.674 \times 10^{-11}$ m$^3$ kg$^{-1}$ s$^{-1}$ es la constante gravitacional de Newton y $\sigma$ es la masa por unidad de área de la hoja metálica. Escriba un programa que calcule y grafique $F_z$ en función de $z$ desde $z=0$ hasta $z=10$m. Analice y discuta el gráfico.

```python

#https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.dblquad.htmlimport numpy as np (ayuda)
import matplotlib.pyplot as plt
from scipy.integrate import dblquad
import numpy as np
# Constantes del problema
G = 6.674e-11  # m^3 kg^-1 s^-1
L = 10.0  # Longitud de cada lado de la hoja en metros
sigma = 100  # Masa por unidad de área de la hoja en kg/m^2
m1 = 1.0  # Masa puntual en kg

# Función a integrar
def integrand(y, x, z):
    return 1 / (x**2 + y**2 + z**2)

# Función que calcula la fuerza gravitacional en la dirección z
def calcular_fuerza_z(z):
    # Definir límites de integración para x e y
    x_min, x_max = -L/2, L/2
    y_min, y_max = -L/2, L/2

    # Realizar la integral
    result, _ = dblquad(integrand, x_min, x_max, lambda x: y_min, lambda x: y_max, args=(z,)) #dblquad para integrales dobles

    # Calcular la fuerza gravitacional
    fuerza_z = G * sigma * m1 * result

    return fuerza_z

# Crear un conjunto de puntos z para evaluar la función
z_values = np.linspace(0, 10, 100)

# Calcular la fuerza gravitacional para cada valor de z
fuerza_z_values = [calcular_fuerza_z(z) for z in z_values]

# Graficar los resultados
plt.plot(z_values, fuerza_z_values, color='blue')
plt.title('Fuerza Gravitacional en función de z')
plt.xlabel('z (m)')
plt.ylabel('Fuerza Gravitacional (N)')
plt.grid(True)
plt.show()
```

La interpretación de este comportamiento en términos físicos podría ser que la hoja metálica ejerce una fuerza gravitacional significativa en objetos cercanos a ella, pero a medida que se alejan, esa influencia disminuye rápidamente y eventualmente se vuelve insignificante a distancias grandes.
